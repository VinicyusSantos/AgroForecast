# **COMEX (EXPORTAÇÕES)**

**ano_mes**

Mês de referência da exportação.
Exemplo:
2020-05-01
significa:
Exportações brasileiras de soja de Maio/2020.

**volume_kg**

Quantidade total exportada de soja em quilogramas.
Exemplo:
10.000.000 kg

10 mil toneladas

**volume_ton**

Quantidade total exportada convertida para toneladas.
Fórmula:
volume_ton = volume_kg / 1000

**valor_fob_usd**

Valor total exportado em dólares FOB.

FOB significa:
Free On Board
ou seja:
valor da mercadoria embarcada no porto brasileiro

**preco_implicito_usd_ton**

Preço médio implícito da soja exportada.
Fórmula:
valor_fob_usd / volume_ton
Exemplo:
US$ 400 / tonelada

```python
exportacoes_comex = pd.read_parquet("fact_exports_monthly.parquet")
exportacoes_comex.head()
```

# **CONAB**

**safra**

Identificador da safra agrícola.
Exemplo:
2024/25

**area_mil_ha**

Área plantada de soja em milhares de hectares.
Exemplo:
45000
45 milhões de hectares

**produtividade_kg_ha**

Quantidade produzida por hectare plantado.
Exemplo:
3500 kg/ha

**producao_mil_ton**

Produção total da safra em milhares de toneladas.
Exemplo:
160000
160 milhões de toneladas

**safra_inicio**

Ano de início da safra.
Exemplo:
2024

**safra_fim**

Ano de término da safra.
Exemplo:
2025
