```python
import pandas as pd
```

```python
df = pd.read_csv("dolar_ptax_raw.csv")
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
df.info()
```

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

## Exportando dados diarios

```python
df.to_parquet('df_daily-dolarptax.parquet', index=False)
```

## Dataset semanal

```python
df = df.set_index('data')
df.head()
```

```python
df_weekly = (
    df
    .resample("W-FRI")
    .agg(
        dolar_medio_semana=("valor", "mean"),
        dolar_fechamento_semana=("valor", "last"),
        dolar_max_semana=("valor", "max"),
        dolar_min_semana=("valor", "min"),
    )
)
```

```python
df_weekly.head()
```

```python
df_weekly.to_parquet('df_weekly-dolarptax.parquet', index=True)
```
