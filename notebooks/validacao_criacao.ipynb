# **Dataset**

```python
START_DATE = "2016-06-10"
END_DATE = "2026-06-05"

cepea = cepea[
    cepea["semana"].between(START_DATE, END_DATE)
]

weather = weather[
    weather["semana"].between(START_DATE, END_DATE)
]

calendario = calendario[
    (calendario.index >= START_DATE) & (calendario.index <= END_DATE)
]
```

```python
cepea_paranagua = cepea[cepea['indicador'] == 'paranagua']
cepea_paranagua.head()
```

```python
dolar.head()
```

```python
conab.head()
```

```python
comex.head()
```

```python
weather.head()
```

```python
calendario.head()
```

```python
mart = (
    cepea_paranagua
    .merge(
        dolar.reset_index(),
        left_on="semana",
        right_on="data",
        how="inner"
    )
    .merge(
        weather,
        on="semana",
        how="inner"
    )
    .merge(
        calendario.reset_index(),
        left_on="semana",
        right_on="data_semana",
        how="inner"
    )
)
```

```python
mart.head()
```

```python
mart.info()
```

```python
mart = mart.drop(columns=['data', 'data_semana', 'indicador', 'ano', 'stress_chuva', 'stress_calor', 'stress_dias_secos', 'preco_rs_max', 'preco_rs_min', 'preco_usd_max', 'preco_usd_min', 'qtd_dias_semana', 'dolar_max_semana', 'dolar_min_semana','temp_max_soja_1w','temp_min_soja_1w','umidade_soja_1w', 'n_pontos', 'dias_calor_extremo_soja_1w'])
```

```python
mart.head()
```

```python
mart.info()
```

```python
mart.duplicated(['semana']).sum()
```

```python
mart_v2 = mart.copy()

mart_v2['ano_mes_semana'] = mart_v2['semana'].dt.to_period("M").dt.to_timestamp()

mart_v2['ano_mes_comex_ref'] = mart_v2['ano_mes_semana'] - pd.DateOffset(months=1)

comex_selected = comex[
    [
        "ano_mes",
        "volume_ton",
        "valor_fob_usd",
        "preco_implicito_usd_ton"
    ]
].copy()

comex_selected = comex_selected.rename(
    columns={
        "ano_mes": "ano_mes_comex_ref",
        "volume_ton": "export_volume_ton_lag_1m",
        "valor_fob_usd": "export_valor_fob_usd_lag_1m",
        "preco_implicito_usd_ton": "export_preco_implicito_usd_ton_lag_1m"
    }
)

mart_v2 = mart_v2.merge(
    comex_selected,
    on="ano_mes_comex_ref",
    how="left"
)

def get_safra_inicio(data):
    ano = data.year
    mes = data.month

    if mes >= 9:
        return ano
    else:
        return ano - 1


def format_safra(safra_inicio):
    safra_fim = safra_inicio + 1
    return f"{safra_inicio}/{str(safra_fim)[-2:]}"


mart_v2["safra_inicio_ref"] = mart_v2["semana"].apply(get_safra_inicio)
mart_v2["safra_ref"] = mart_v2["safra_inicio_ref"].apply(format_safra)

conab_selected = conab[
    [
        "safra",
        "area_mil_ha",
        "produtividade_kg_ha",
        "producao_mil_ton"
    ]
].copy()

conab_selected = conab_selected.rename(
    columns={
        "safra": "safra_ref",
        "area_mil_ha": "conab_area_mil_ha",
        "produtividade_kg_ha": "conab_produtividade_kg_ha",
        "producao_mil_ton": "conab_producao_mil_ton"
    }
)

mart_v2 = mart_v2.merge(
    conab_selected,
    on="safra_ref",
    how="left"
)

cols_to_drop = [
    "ano_mes_semana",
    "ano_mes_comex_ref",
    "safra_inicio_ref",
]

mart_v2 = mart_v2.drop(columns=cols_to_drop)


print("Shape:", mart_v2.shape)
print("Período:", mart_v2["semana"].min(), "até", mart_v2["semana"].max())
print("Duplicatas semana:", mart_v2.duplicated(["semana"]).sum())

print("\nNulos:")
print(mart_v2.isna().sum().sort_values(ascending=False))
```
